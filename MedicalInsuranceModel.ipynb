```python
smoker_codes = {'no': 0, 'yes':1}
medical_df['smoker_codes'] = medical_df.smoker.map(smoker_codes)
```

```python
medical_df
```

```python
sex_codes = {'female':0, 'male': 1}
medical_df['sex_codes'] = medical_df.sex.map(sex_codes)
```

```python
medical_df
```

```python
# one hot encoding
from sklearn import preprocessing
enc = preprocessing.OneHotEncoder()
enc.fit(medical_df[['region']])
print(enc.categories_)
one_hot = enc.transform(medical_df[['region']]).toarray()
print(one_hot)
```

```python
medical_df[['northeast', 'northwest', 'southeast', 'southwest']] = one_hot
```

```python
medical_df
```

```python
# scaling
from sklearn.preprocessing import StandardScaler

numeric_cols = ['age', 'bmi']
scalar = StandardScaler()
scalar.fit(medical_df[numeric_cols])

scaled_inputs = scalar.transform(medical_df[numeric_cols])
scaled_inputs
```

```python
# categorical or numerical col
cat_cols = ['smoker_codes', 'sex_codes', 'northeast', 'northwest', 'southeast', 'southwest']
categorical_data = medical_df[cat_cols].values
categorical_data
```

```python
import numpy as np

def rmse(target, prediction):
    return np.sqrt(np.mean(np.square(target - prediction)))
```

```python
# model
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

inputs = np.concatenate((scaled_inputs, categorical_data), axis=1)
targets = medical_df.charges

inputs_train, inputs_test, targets_train, targets_test = train_test_split(inputs, targets, test_size=0.1)

model = LinearRegression()

model.fit(inputs_train, targets_train)

prediction_test = model.predict(inputs_test)

loss = rmse(targets_test, prediction_test)
print("RMSE Loss", loss)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import math

print("R² Score:", r2_score(targets_test, prediction_test))
print("MAE:", mean_absolute_error(targets_test, prediction_test))
print("MSE:", mean_squared_error(targets_test, prediction_test))
print("Root Mean Squared Error: ",math.sqrt(mean_squared_error(targets_test, prediction_test)))
```
